# we can create a function that returns name of team against which a particular batsman scored the most


```python
def highest_runs_scored(batsman_name):
    vk = deliveries[deliveries['batsman']==batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending= False).head(1).index[0]
    
```

```python
highest_runs_scored('DA Warner')
```

```python
highest_runs_scored('G Gambhir')
```

```python
highest_runs_scored('MS Dhoni')
```

# isin() Function 


<div>isin()</div>
<div>Finding the most destructive death over batsman in the IPL history!</div>
<div>Using Strike Rate -> (Number of runs/Number of balls)*100</div>
<div>Criteria -> Min batsman 200 balls in the overs 16-20</div>

```python
#getting the death overs 

death_over = deliveries[deliveries['over']>15]
death_over.head()
```

```python
# finding the number of balls played by each batsman who are also present in the deathover
req_batsman = death_over.groupby('batsman')['batsman_runs'].count()
req_batsman
# now we do masking for batsman playing 200+ balls
mask = req_batsman >200
req_batsman[mask]
```

```python
# now we extract all batsman names 
req_batsman[mask].index   #this gives us a list 
```

<h1>tolist()</h1> 

```python
#we can make a list in proper format using tolist() 
req_list=req_batsman[mask].index.tolist()
req_list
```

```python
# now for the strike rate we req runs scored by these batsmen and number of balls played by them 
# we can use deliveries data to find runs scored but it also has other batsmen data 
# here we can use isin() function 

```

<h1>isin() requires a list input so we have created a list of batsmen above</h1>

```python
final = deliveries[deliveries['batsman'].isin(req_list)]
final
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
strike_rate = (runs/balls)*100
```

# hence we have the deadliest batsmen in IPL

```python
strike_rate.sort_values(ascending = False).head()
```
